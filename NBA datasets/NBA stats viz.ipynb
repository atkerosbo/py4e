```python
import pandas as pd
```

```python
df = pd.read_csv('shot_logs.csv')
```

```python
df.head()
```

```python
df.groupby('CLOSEST_DEFENDER').get_group('Bogdanovic, Bojan')
```

```python
df_shots_made = df.groupby('SHOT_RESULT').get_group('made').copy()
```

```python
df_shots_made.head()
```

```python
df_shots_made.groupby('player_name').get_group('kemba walker')
```

```python
df_shots_made_kemba = df_shots_made.groupby('player_name').get_group('kemba walker').copy()
```

```python
df_shots_made_kemba.to_csv('kembamade.csv', header=True, index=True)
```

```python
df_shots_made['player_name'] = df_shots_made['player_name'].map(lambda x: x.replace(" ", ""))
```

```python
df_shots_made.groupby('player_name').apply(lambda x: x.to_csv(f'{x.name}.csv'))
```

```python
for i in df_shots_made['player_name']:
    df_player = df_shots_made.groupby('player_name').get_group(i).copy()
    df_player.to_csv('[i].csv')
```
